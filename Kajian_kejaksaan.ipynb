```python
import pandas as pd

df = pd.read_csv('/content/hasil_scraping (1).csv')
df
```

```python
df.duplicated().sum()
```

```python
df.isna().sum()
```

```python
df.drop(columns=['Kategori'], axis=1, inplace=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Mengubah kolom Date menjadi format datetime
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Menghitung jumlah berita per tanggal
jumlah_berita = (
    df.groupby(df["Date"].dt.date)
      .size()
      .reset_index(name="Jumlah Berita")
)

# Mengurutkan berdasarkan tanggal
jumlah_berita = jumlah_berita.sort_values("Date")

# Menampilkan hasil
print(jumlah_berita)

bars = plt.bar(
    jumlah_berita["Date"].astype(str),
    jumlah_berita["Jumlah Berita"],
    color="steelblue"
)

# Menambahkan jumlah kasus di atas setiap batang
for bar in bars:
    tinggi = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        tinggi + 0.1,              # posisi sedikit di atas batang
        f"{int(tinggi)}",          # menampilkan jumlah kasus
        ha="center",
        va="bottom",
        fontsize=10,
        fontweight="bold"
    )

# Garis tren
plt.plot(
    jumlah_berita["Date"].astype(str),
    jumlah_berita["Jumlah Berita"],
    marker="o",
    linewidth=2,
    color="red",
    label="Tren"
)

plt.title("Jumlah Artikel Berita tentang Brigjen Lalu")
plt.xlabel("Tanggal")
plt.ylabel("Jumlah Berita")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
